```python
anti_toxicity_scores = []
pro_toxicity_scores = []
```

```python
import csv


file_path = "Covid19-sentences.csv"
with open(file_path, newline='') as csvfile:
    reader = csv.DictReader(csvfile)

    column_names = reader.fieldnames
    print("Column names:", column_names)

    for row in reader:
        analyze_request = {
            'comment': {'text': row['\ufeffanti_vaccine_text']},
            'requestedAttributes': {'TOXICITY': {}}
        }

        response = client.comments().analyze(body=analyze_request).execute()


        toxicity_score = response['attributeScores']['TOXICITY']['summaryScore']['value']
        anti_toxicity_scores.append(toxicity_score)

        print("Sentence:", row['\ufeffanti_vaccine_text'])
        print("Toxicity Score:", toxicity_score,"\n")

    print("Anit-Vaccine Sentences Complete")

```

```python
file_path = "Covid19-sentences.csv"
with open(file_path, newline='') as csvfile:
    reader = csv.DictReader(csvfile)

    column_names = reader.fieldnames
    print("Column names:", column_names)

    for row in reader:
        analyze_request = {
            'comment': {'text': row['pro_vaccine_text']},
            'requestedAttributes': {'TOXICITY': {}}
        }

        response = client.comments().analyze(body=analyze_request).execute()


        toxicity_score = response['attributeScores']['TOXICITY']['summaryScore']['value']
        pro_toxicity_scores.append(toxicity_score)

        print("Sentence:", row['pro_vaccine_text'])
        print("Toxicity Score:", toxicity_score,"\n")

    print("Pro-Vaccine Sentences Complete")
```

```python

average_anti_score = sum(anti_toxicity_scores)/len(anti_toxicity_scores)
print("This is the average toxicity score of the anti-vaccine sentences:", average_anti_score)

average_pro_score = sum(pro_toxicity_scores)/len(pro_toxicity_scores)
print("This is the average toxicity score of the pro-vaccine sentences:", average_pro_score)
```
